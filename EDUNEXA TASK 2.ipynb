# using Matplotlib

```python
import matplotlib.pyplot as plt

# Data
x = [1, 2, 3, 4, 5]
y = [10, 15, 13, 18, 20]

# Create a basic line plot
plt.plot(x, y)

# Display the plot
plt.show()
```

```python
#Customizing Plots: Exploring customization options like labels, colors, and line stylesÂ¶
plt.plot(x, y, color='blue', linestyle='dashed', marker='o')
plt.title('Customized Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid(True)
plt.show()
```

```python
#Adding Titles and Labels: Using plt.title(), plt.xlabel(), and plt.ylabel() to enhance your plots
plt.plot(x, y)
plt.title('Line Plot Example')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

```python
#Plotting Multiple Data Sets: Plotting multiple lines on the same graphÂ¶
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y1 = [10, 15, 13, 18, 20]
y2 = [5, 8, 7, 12, 15]

plt.plot(x, y1, label='Line 1')
plt.plot(x, y2, label='Line 2')

plt.title('Multiple Lines on One Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

plt.legend()  # Show legends
plt.show()
```

```python
#Legends and Annotations: Adding legends and annotations to your plots
plt.plot(x, y1, label='Line 1')
plt.plot(x, y2, label='Line 2')

plt.title('Multiple Lines with Legends and Annotations')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

plt.legend()  # Show legends

# Add annotations
plt.annotate('Point A', xy=(2, 15), xytext=(3, 17),
             arrowprops=dict(arrowstyle='->'))
plt.text(4, 8, 'Important Point', fontsize=12, color='red')

plt.show()
```

# Different Types of Plots

```python
#Scatter Plots: Creating scatter plots with plt.scatter()
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [10, 15, 13, 18, 20]

plt.scatter(x, y, color='blue', marker='o', label='Data Points')
plt.title('Scatter Plot Example')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.show()
```

```python
#Bar Charts: Using plt.bar() for bar chartsÂ¶
categories = ['Category A', 'Category B', 'Category C']
values = [15, 30, 25]

plt.bar(categories, values, color='pink')
plt.title('Bar Chart Example')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.show()
```

```python
#Histograms: Generating histograms with plt.hist()Â¶
data = [1, 2, 2, 3, 3, 3, 4, 4, 5, 5, 5, 5]

plt.hist(data, bins=5, edgecolor='k', color='skyblue')
plt.title('Histogram Example')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.show()
```

```python
#Pie Charts: Creating pie charts using plt.pie()
labels = ['Category A', 'Category B', 'Category C']
sizes = [15, 30, 25]
colors = ['gold', 'lightcoral', 'lightskyblue']

plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)
plt.title('Pie Chart Example')
plt.show()
```

```python
#Box Plots: Visualizing data distribution with plt.boxplot()Â¶
data = [85, 90, 70, 100, 110, 95, 75, 120]

plt.boxplot(data, vert=False)
plt.title('Box Plot Example')
plt.xlabel('Values')
plt.show()
```

# Subplots and Layouts

```python
#Subplots: Dividing a figure into multiple subplots with plt.subplot()
import matplotlib.pyplot as plt

# Create a figure with 2x2 subplots
plt.figure(figsize=(10, 6))  # Set the overall figure size
plt.subplot(2, 2, 1)  # The first subplot
plt.plot([1, 2, 3, 4], [10, 15, 13, 18])
plt.title('Subplot 1')

plt.subplot(2, 2, 2)  # The second subplot
plt.scatter([1, 2, 3, 4], [10, 15, 13, 18], color='red')
plt.title('Subplot 2')

plt.subplot(2, 2, 3)  # The third subplot
plt.bar(['A', 'B', 'C'], [15, 30, 25], color='green')
plt.title('Subplot 3')
plt.subplot(2, 2, 4)  # The fourth subplot
plt.hist([1, 2, 2, 3, 3, 3, 4, 4, 5, 5, 5, 5], bins=5, edgecolor='k', color='skyblue')
plt.title('Subplot 4')

plt.suptitle('Subplots Example')  # Set a title for the entire figure
plt.tight_layout()  # Ensure proper spacing
plt.show()
```

# Advanced Plot Types

```python
#3D Plots: Creating 3D plots with mpl_toolkits.mplot3d
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Create a meshgrid
x = np.linspace(-5, 5, 50)
y = np.linspace(-5, 5, 50)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))

# Create a 3D figure
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z)

# Customize the plot
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')
ax.set_title('3D Surface Plot')

plt.show()
```

```python
#Heatmaps: Generating heatmaps for displaying data matrices
import numpy as np
import matplotlib.pyplot as plt

# Create a sample data matrix
data = np.random.rand(10, 10)

# Create a heatmap
plt.imshow(data, cmap='YlGnBu', interpolation='nearest')
plt.colorbar()
plt.title('Heatmap Example')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#1. Histograms with sns.histplot():
import seaborn as sns
import matplotlib.pyplot as plt

# Sample data
data = sns.load_dataset("iris")

# Create a histogram of sepal length (a single variable)
sns.histplot(data=data, x="sepal_length", kde=True, color="skyblue")

# Add labels and title
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Frequency")
plt.title("Histogram of Sepal Length")

# Display the plot
plt.show()
```

```python
#2. Scatter Plots with sns.scatterplot():
import seaborn as sns
import matplotlib.pyplot as plt

# Sample data
data = sns.load_dataset("iris")

# Create a scatter plot of sepal length vs. sepal width
sns.scatterplot(data=data, x="sepal_length", y="sepal_width", hue="species", palette="viridis")

# Add labels and title
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Sepal Width (cm)")
plt.title("Scatter Plot of Sepal Length vs. Sepal Width")

# Display the plot
plt.show()
```

```python
#3. Box Plots with sns.boxplot():
import seaborn as sns
import matplotlib.pyplot as plt

# Sample data
data = sns.load_dataset("iris")

# Create a box plot of petal length by species
sns.boxplot(data=data, x="species", y="petal_length", palette="Set3")

# Add labels and title
plt.xlabel("Species")
plt.ylabel("Petal Length (cm)")
plt.title("Box Plot of Petal Length by Species")

# Display the plot
plt.show()
```
